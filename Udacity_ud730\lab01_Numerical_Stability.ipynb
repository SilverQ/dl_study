* Set a variable to 1 billion (109). Then add 10-6 to it a million (106) times.
* Now subtract 1 billion from it. What is the result?

```
a = 1000000000
b = 0.000001

# print(a+b)

for i in range(1000000):
  a += b
#   if i % 100000 == 0:
  if i < 100:
    print(i, a)

a -= 1000000000
print(i, a)
```

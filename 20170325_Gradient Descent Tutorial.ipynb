위의 링크를 참고하여, 원하는 함수, 러닝 레이트, 등을 변경함에 따른 테스트 결과를 볼 수 있음
러닝 레이트를 너무 크게 잡으면(0.1) 수렴하지 않고, 발산하는 것을 관찰 가능
러닝 레이트를 작게 잡을 수록 iteration 횟수는 증가

```python
x_old = 0
iter = 0
```

```python
x_new = 6 # The algorithm starts at x=6
```

```python
eps = 0.001 # step size
precision = 0.00001
```

eps = 0.01, precision = 0.00001
Local minimum occurs at 2.24996460743, 70 times iterated

eps = 0.001, precision = 0.000001
Local minimum occurs at 2.25047815041, 358 times iterated

```python
def f_prime(x):
    return 4 * x**3 - 9 * x**2
```

```python
while abs(x_new - x_old) > precision:
    iter = iter + 1
    x_old = x_new
    x_new = x_old - eps * f_prime(x_old)
    print(x_old, f_prime(x_old), x_new)
```

```python
print ('Local minimum occurs at %s, %s times iterated' % (x_new, iter))
```
